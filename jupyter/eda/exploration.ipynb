### Data visualisation

```python
# Variable categorielles
categorical_columns = data.nunique()[data.nunique() < 5].keys().to_list()
categorical_columns
```

```python
# Construction de graphique à barres interactive
def bar_plot(Variable):
    plt.rcParams["figure.figsize"] = (7, 4)
    sns.set_style("darkgrid")
    data[Variable].value_counts(normalize=True).plot(kind='bar', color='violet')
    plt.ylabel('proportion', size=14)
    plt.title('Distribution of ' + str(Variable), size=16)
    return plt.show()
```

```python
# Interact, barplot
interact(bar_plot, Variable = categorical_columns)
```

```python
# Verifier le type de 'TotalCharges'
data['TotalCharges'].sample(10)
```

```python
# Variables quantitatives
numerical_columns = ['tenure', 'MonthlyCharges', 'TotalCharges']

# Convertire le type de 'TotalCharges' en float
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
# Construction d'histogrammes interactive
def hist_plot(Variable):
    plt.rcParams["figure.figsize"] = (7, 4)
    sns.set_style("darkgrid")
    sns.distplot(data[Variable].dropna(), kde=False)
    plt.title('Histogram of ' + str(Variable), size=16)
    return plt.show()
```

```python
# Interact, histogramme
interact(hist_plot, Variable = numerical_columns);
```

```python
# Coefficient d'asymetrie 'TotalCharges'
data['TotalCharges'].skew()
```

```python
# Construction de boîtes à moustaches interactive
def box_plot(Variable):
    plt.rcParams["figure.figsize"] = (7, 4)
    sns.set_style("darkgrid")
    sns.boxplot(y=data[Variable].dropna(), color='yellow')
    plt.title('Boxplot of ' + str(Variable), size=16)
    return plt.show()
```

```python
# Interact, boites à moustaches 
interact(box_plot, Variable = numerical_columns);
```

### Data preprocessing

```python
# Gestion des valeurs manquantes
df_copy = data.copy()

# Valeurs manquantes par variable
df_copy.isna().sum()
```

```python
# Suppression vm
df_copy.dropna(inplace = True)
```

```python
# Encodage variable binaire gender
df_copy['gender'] = df_copy['gender'].apply(lambda row: 1 if row == 'Male' else 0)

# Liste contenant tout les variables binaire excepter gender
binary_columns = df_copy.drop('gender', axis=1).nunique()[df_copy.drop('gender', axis=1).nunique() < 3].keys().to_list()
```

```python
# Encodage des columns categorique
for column in binary_columns:
  df_copy[column] = df_copy[column].apply(lambda row: 1 if row == 'Yes' else 0)
```

```python
# Liste de variables catégorielles avec plus de 2 modalités
remaining = df_copy[categorical_columns].nunique()[df_copy[categorical_columns].nunique() > 2].keys().to_list()

# Encodage avec get_dummies
remaining_dummies = pd.get_dummies(df_copy[remaining], columns=remaining, drop_first=True)
```

```python
# Nouvelle dataframe
df_copy = pd.concat([df_copy['gender'], df_copy[binary_columns], remaining_dummies, df_copy[numerical_columns]], axis = 1)

# Affichage de la DataFrame
df_copy.head(10)
```

```python
# Resolution assymetrie 'TotalCharges'
df_copy['TotalCharges'] = np.sqrt(df_copy['TotalCharges'])

# Histogramme de la variables transformée
sns.distplot(df_copy['TotalCharges'], kde = False);
```

```python
# Coéfficient d'asymétrie 'TotalCharge'
print("Avant résolution:", data['TotalCharges'].skew(), "\n")
print("Après résolution:", df_copy['TotalCharges'].skew())
```

```python
Features = df_copy.drop('Churn', axis =1)
Labels = df_copy['Churn']
```

```python
# Sauvegarde des data
Features.to_csv('Output/feature_clean.csv', index=False)
Labels.to_csv('Output/labels_clean.csv', index=False)
```
